### Задание № 1.

Создайте абстрактный класс Figure и объявите у него два абстрактных метода calculate_perimeter, и calculate_area
Реализацию методов оставьте пустой.

```python
from abc import ABC, abstractmethod

class Figure(ABC):
    @abstractmethod
    def calculate_perimeter(self):
        pass

    @abstractmethod
    def calculate_area(self):
        pass
```

### Задание № 2.
Наследуясь от абстрактного класса Figure создайте классы Rectangle и Tringle со своими инициализаторами. Реализуйте у них абстрактные методы.

```python
class Rectangle(Figure):
    def __init__(self, a, b):
        if a <= 0 or b <= 0:
            raise ValueError("Стороны должны быть положительны")
        self.a = a
        self.b = b

    def calculate_perimeter(self):
        return 2 * (self.a + self.b)

    def calculate_area(self):
        return self.a * self.b
```

```python
class Tringle(Figure):
    def __init__(self, side1, side2, side3):
        if side1 <= 0 or side2 <= 0 or side3 <= 0:
            raise ValueError("Стороны должны быть положительны")
        if side1 + side2 <= side3 or side1 + side3 <= side2 or side2 + side3 <= side1:
            raise ValueError("Неверное соотношение сторон треугольника")
        self.side1 = side1
        self.side2 = side2
        self.side3 = side3

    def calculate_perimeter(self):
        return self.side1 + self.side2 + self.side3

    def calculate_area(self):
        halfperimeter = self.calculate_perimeter() / 2
        return (halfperimeter * (halfperimeter - self.side1) * (halfperimeter - self.side2) * (halfperimeter - self.side3)) ** 0.5
```

### Задание № 11.
Защитите параметры длин сторон от прямого изменения через точку.

rectangle.a # 2<br>
rectangle.a = 5<br>

```python
class RectangleException(ValueError):
    def __init__(self, errortext="Длины сторон прямоугольника должны быть положительными!"):
        super().__init__(errortext)


class Rectangle(Figure):
    def __init__(self, a, b):
        if a <= 0 or b <= 0:
            raise RectangleException()

        self._a = a
        self._b = b

        self._perimeter = None
        self._area = None

    @property
    def a(self):
        return self._a

    @property
    def b(self):
        return self._b

    def calculate_perimeter(self):
        if self._perimeter is None:
            self._perimeter = 2 * (self.a + self.b)
        return self._perimeter

    def calculate_area(self):
        if self._area is None:
            self._area = self.a * self.b
        return self._area

    def __repr__(self):
        return f"Прямоугольник со сторонами {self.a} и {self.b}"

    def __eq__(self, other):
        return self.calculate_area() == other.calculate_area()

    def __gt__(self, other):
        return self.calculate_area() > other.calculate_area()

    def __lt__(self, other):
        return self.calculate_area() < other.calculate_area()

    def __le__(self, other):
        return self.calculate_area() <= other.calculate_area()

    def __ge__(self, other):
        return self.calculate_area() >= other.calculate_area()

    def __add__(self, other):
        new_a = self.a + other.a
        new_b = self.b + other.b
        return Rectangle(new_a, new_b)

    def __sub__(self, other):
        new_a = self.a - other.a
        new_b = self.b - other.b
        return Rectangle(new_a, new_b)
```
